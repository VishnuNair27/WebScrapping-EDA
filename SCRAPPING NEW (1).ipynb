```python
Brand=[]
Model=[]
Color=[]
Type=[]
Rating=[]
No_rating=[]
Bought=[]
Discounted_price=[]
Original_price=[]
Discount=[]

for j in range(1,50):
    url=f'https://www.amazon.in/s?i=electronics&rh=n%3A976419031%2Cn%3A29561348031%2Cn%3A1388921031%2Cn%3A14146389031&dc&page={j}&qid=1748416286&rnid=1388921031&xpid=z1T9uJpqzN1iS&ref=sr_pg_{j}'
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text)

    section=soup.find_all('div',class_='a-section a-spacing-base')
    for i in section:
        info=i.find('a',class_="a-link-normal s-line-clamp-4 s-link-style a-text-normal")
        info=info.text
        
 ##Brand---------------------------------------------------------------------
        brand=re.findall('^\w+',info)
        if brand:
            Brand.append(brand[0])
        else:
            Brand.append(np.nan)
            
 ##Model---------------------------------------------------------------------
        
        model=re.findall( r'^(.*?)\s+(?:Bluetooth|Wired|Truly|in-Ear|in Ear|Wireless|Earbuds|Headphones|Earphones|TWS)',info)
        if model:
            Model.append(model[0])
        else:
            Model.append(np.nan)
 ##color-----------------------------------------------------------------------   
        color=re.findall('\(([a-zA-Z]+ ?[a-zA-Z]+)',info)
        if color:
            Color.append(color[0])
        else:
            Color.append(np.nan)
  ##type-----------------------------------------------------------------------          
        type_of=re.findall('(Wireless|Wired)',info)
        if type_of:
            Type.append(type_of[0])
        else:
            Type.append(np.nan)
            
   ##rating info---------------------------------------------------------------- 
        Rating_info=i.find('div',class_='a-section a-spacing-none a-spacing-top-micro')
        Rating_info=Rating_info.text
        rating=re.findall('^\d+.\d',Rating_info)
        if rating:
            Rating.append(rating[0])
        else:
            Rating.append(np.nan)
        num_rating=re.findall('stars ([\d,]+)',Rating_info)
        if num_rating:
            No_rating.append(num_rating[0])
        else:
            No_rating.append(np.nan)
            
    ##bought-------------------------------------------------------------------        
        bought=re.findall('[\dK+]+ bought \w.+',Rating_info)
        if bought:
            Bought.append(bought[0])
        else:
            Bought.append(np.nan)
   ##price-------------------------------------------------------------------------         
        price_info=i.find('div',class_='a-row a-size-base a-color-base')
        price_info=price_info.text
        
   ##discount------------------------------------------------------------------     
        price=re.findall('page(₹[\d+,]+)',price_info)
        if price:
            Discounted_price.append(price[0])
        else:
            Discounted_price.append(np.nan)

   ##org price-----------------------------------------------------------------------     
        ori_price=re.findall('M.R.P: (₹[,\d+]+)₹',price_info)
        if ori_price:
            Original_price.append(ori_price[0])
        else:
            Original_price.append(np.nan)
  ##discount %--------------------------------------------------------------------------          
        discount=re.findall('([0-9%]+) off',price_info)
        if discount:
            Discount.append(discount[0])
        else:
            Discount.append(np.nan)
```

```python
## to create dataframe:

data={
    'Brand':Brand,
      'Model':Model,
      'Color':Color,
      'Type':Type,
      'Rating':Rating,
      'No_rating':No_rating,
      'Bought':Bought,
      'Discounted_price':Discounted_price,
      'Original_price':Original_price,
      'Discount':Discount
     }
```
